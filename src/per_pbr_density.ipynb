```python


import seaborn as sns
import matplotlib.pyplot as plt
from pykrx import stock

def show_per_pbr(market):
    market = stock.get_market_fundamental("20220603", market=market)
    market_PER = market["PER"]
    market_PBR = market["PBR"]
    
    # 등록되지 않은 PER, 값이 너무 큰 PER 제거
    market_PER = market_PER[market_PER != 0]
    market_PER = market_PER[market_PER < 50] 
    
    # 등록되지 않은 PBR, 값이 너무 큰 PBR 제거
    market_PBR = market_PBR[market_PBR != 0]
    market_PBR = market_PBR[market_PBR < 5] 
    
    sns.kdeplot(market_PER).set_title("PER")
    plt.figure()
    sns.kdeplot(market_PBR).set_title("PBR")
    plt.figure()
    

show_per_pbr("ALL")
# show_per_pbr("KOSPI")
# show_per_pbr("KOSDAK")


```
